```python
#NÜMERİK BİR DEĞİŞKENİ KATEGORİKLEŞTİRME
#SaleCheckInDayDiff değişkenini kategorik bir değişkene çeviriniz.
#pandastaki cut fonksiyonu ile veriyi istenilen aralıklara göre etiketleyebiliriz
#aralıklar --> ‘0_7’, ‘7_30', ‘30_90', ‘90_max’
#labels --> "Last Minuters", "Potential Planners", "Planners", "Early Bookers“
#EB_Score isimli yeni bir sütuna yazdır.
bins= [-1,7,30,90,df["SaleCheckInDayDiff"].max()]
labels= ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]

df["EB_Score"]=pd.cut(x=df["SaleCheckInDayDiff"], bins=bins, labels=labels)
df
```

```python
#Şehir-Concept-EB Score, Şehir-Concept- Sezon, Şehir-Concept-CInDay kırılımında ortalama ödenen ücret ve yapılan işlem sayısı cinsinden inceleyiniz
#pivot table mantığı ile bakalım --> groupby
df.groupby(["SaleCityName","ConceptName", "EB_Score"]).agg({"Price":["mean", "count"]})
```

```python
df.groupby(["SaleCityName","ConceptName", "Seasons"]).agg({"Price":["mean", "count"]})
```

```python
#City-Concept-Season kırılımının çıktısını PRICE'a göre sıralayınız.
#Elde ettiğiniz çıktıyı agg_df olarak kaydediniz.
agg_df=df.groupby(["SaleCityName","ConceptName", "Seasons"]).agg({"Price":"mean"})
#azalan sırada yapmak için ascending = False olarak belirtilmeli
agg_df.sort_values(by=["Price"], ascending=False)
```

```python
#Indekste yer alan isimleri değişken ismine çeviriniz.
agg_df.reset_index(inplace=True)
agg_df
```

```python
#Yeni seviye tabanlı müşterileri (persona) tanımlayınız.
#Yeni seviye tabanlı satışları tanımlayınız ve veri setine değişken olarak ekleyiniz.
#Yeni eklenecek değişkenin adı: sales_level_based
#Önceki soruda elde edeceğiniz çıktıdaki gözlemleri bir araya getirerek sales_level_based değişkenini oluşturmanız gerekmektedir.

agg_df["Sales_Level_Based"]=agg_df["SaleCityName"]+ " "+ agg_df["ConceptName"] + " "+ agg_df["Seasons"]
agg_df.head()
```

```python
agg_df["Sales_Level_Based"]=agg_df[["SaleCityName","ConceptName","Seasons"]].agg(lambda x: '_'.join(x).upper(),axis=1)
agg_df.head()
```

```python
#Yeni müşterileri (personaları) segmentlere ayırınız.
#Yeni personaları PRICE’a göre 4 segmente ayırınız.
#Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df’e ekleyiniz.
#Segmentleri betimleyiniz (Segmentlere göre group by yapıp price mean, max, sum’larını alınız).
agg_df["Segment"] = pd.qcut(agg_df["Price"], 4, labels=["D", "C", "B", "A"])
agg_df
```
