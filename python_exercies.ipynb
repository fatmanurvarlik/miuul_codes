```python
#Argüman olarak bir liste alan, listenin içerisindeki tek ve çift sayıları ayrı listelere atayan ve bu listeleri return eden fonksiyon yazınız.
l = [2,13,18,93,22]

def func(l):
  even_list=[]
  odd_list=[]
  for number in l:
    if number % 2 == 0:
      even_list.append(number)
    else:
      odd_list.append(number)
  return even_list, odd_list

func(l)
```

```python
"""Görev 6: Aşağıda verilen listede mühendislik ve tıp fakülterinde dereceye giren öğrencilerin isimleri
bulunmaktadır. Sırasıyla ilk üç öğrenci mühendislik fakültesinin başarı sırasını temsil ederken son üç öğrenci de
tıp fakültesi öğrenci sırasına aittir. Enumarate kullanarak öğrenci derecelerini fakülte özelinde yazdırınız."""

ogrenciler=["ali","veli","ayşe","talat","zeynep","ece"]

for index, isim in enumerate(ogrenciler,1):
  if index in [1,2,3]:
    print("mühendislik fakültesi", index, ".", "öğrenci:", isim)
  else:
    print("tıp fakültesi", index,  ".", "öğrenci:", isim)
```

```python
#görev 7: Aşağıda 3 adet liste verilmiştir. Listelerde sırası ile bir dersin kodu, kredisi ve kontenjan bilgileri yer almaktadır. Zip kullanarak ders bilgilerini bastırınız.
ders_kodu=["CMP1005","PSY1001","HUK1005","SEN2204"]
kredi=[3,4,2,4]
kontenjan=[30,75,150,25]
list(zip(ders_kodu,kredi,kontenjan))
#zipin içinde hangi sırayla yazdıysan o şekilde for döngüsüne yazılması gerekiyor aksi takdirde yanlış sıralama ile alıyor
for ders_kodu,kredi,kontenjan in zip(ders_kodu,kredi,kontenjan):
  print(f"kredisi {kredi} olan {ders_kodu} kodlu dersin kontejanı {kontenjan} kişidir" )
```

```python
#Görev 8: Aşağıda 2 adet set verilmiştir. Sizden istenilen eğer 1. küme 2. kümeyi kapsiyor ise ortak elemanlarını
#eğer kapsamıyor ise 2. kümenin 1. kümeden farkını yazdıracak fonksiyonu tanımlamanız beklenmektedir.
kume1=set(["data","python"])
kume2=set(["data","function","qcut","lambda","python","miuul"])

def look_sets(set1,set2):
  if kume1.issuperset(kume2) == 'True':
    print(set1.intersection(set2))
  else:
    print(set1.difference(set2))

look_sets(kume1,kume2) #kume 1 2yi kapsamadığı için sonuç istenilen şekilde çıkmıyor
look_sets(kume2,kume1)
```

```python
import seaborn as sns
df=sns.load_dataset("car_crashes")
df.columns
#List Comprehension yapısı kullanarak car_crashes verisindeki numeric değişkenlerin isimlerini büyük harfe çeviriniz ve başına NUM ekleyiniz.

num_cols = ["NUM_" + col.upper() if df[col].dtype != "O" else col.upper() for col in df.columns ]
num_cols
```

```python
#List Comprehension yapısı kullanarak car_crashes verisinde isminde "no" barındırmayan değişkenlerin isimlerinin sonuna "FLAG" yazınız.

cols = [col.upper() + "_FLAG" if "no" not in col else col.upper() for col in df.columns ]
cols
```

```python
#List Comprehension yapısı kullanarak aşağıda verilen değişken isimlerinden FARKLI olan değişkenlerin isimlerini seçiniz ve yeni bir dataframe oluşturunuz.
og_list=["abbrev","no_previous"]
new_col = [col for col in df.columns if col not in og_list]
df[new_col]
```
